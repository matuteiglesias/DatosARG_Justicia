```python
import pandas as pd
```

```python
balances = pd.read_csv('igj-balances-2018-07.csv')
balances['capital_informado'] = balances['capital_informado'].str.replace(',', '.').astype(float)
balances['balance_year'] = balances.fecha_balance.astype(str).str[:4]
balances['balance_month'] = balances.fecha_balance.astype(str).str[4:6]
balances['balance_day'] = balances.fecha_balance.astype(str).str[6:8]

entidades = pd.read_csv('igj-entidades-2018-07.csv')
```

```python
# balances.groupby('balance_year').count().sort_index(ascending = False)
# balances.loc[balances.balance_year == '2010']
```

```python
vigentes = entidades.loc[entidades.dada_de_baja != 'S']
# vigentes.groupby('descripcion_tipo_societario').count().sort_values(by = 'numero_correlativo', ascending = False)
```

```python
balances.loc[balances.balance_year == '2010'].loc[balances.numero_correlativo.isin(vigentes.numero_correlativo)].groupby('descripcion_tipo_societario')['capital_informado'].sum().sort_values()
```

```python
balances.loc[balances.balance_year == '2010'].loc[balances.numero_correlativo.isin(vigentes.numero_correlativo)].groupby('descripcion_tipo_societario')['capital_informado'].median().sort_values()
```

## Decalraciones juradas

```python
dj = pd.read_csv('./declaraciones-juradas-2017-actualizado-al-2017-08-31.csv')
# dj.dtypes == 'float64'#.sum().sort_values()

# dj.select_dtypes(include='float64').sum().sort_values().index
dj_main = pd.concat([dj.select_dtypes(exclude='float64'), dj[[u'ingresos_no_alcanzados', u'ingresos_no_alcanzados_por_ig',
       u'deducciones_generales', u'total_gastos_c4', u'total_deudas_final',
       u'deudas_inicio', u'gastos_personales', u'diferencia_valuacion',
       u'ingresos_neto_gastos', u'ingresos_trabajos_alquileres_rentas',
       u'ingreso_neto_renta_trabajo_personal_c4', u'total_ingreso_neto_c1234',
       u'total_ingresos_c4', u'total_bienes_final', u'total_bienes_inicio']]], axis = 1)
```

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

pd.options.display.max_rows = 999

dj_main.groupby(['organismo', 'cargo']).agg({'anio': 'count', 'total_bienes_inicio': 'sum', 'apellido_nombre': 'first'}).sort_values(by = 'total_bienes_inicio', ascending = False).head(300)
```
